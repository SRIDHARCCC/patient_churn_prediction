```python
spark.sql("""
CREATE SCHEMA IF NOT EXISTS patient_churn_prediction.log_model
""")

spark.sql("""
CREATE SCHEMA IF NOT EXISTS patient_churn_prediction.patient_data
""")

spark.sql("""
CREATE SCHEMA IF NOT EXISTS patient_churn_prediction.source_data
""")
```
