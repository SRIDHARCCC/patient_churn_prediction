```python
from pyspark.sql import functions as F
from pyspark.sql.functions import to_date, datediff, lit
```

```python
df = spark.table("patient_churn_prediction.patient_data.bronze_patients")
```

```python
silver_df  = df.withColumn(
    "Days_since_interacted",
    datediff(lit("2026-01-25"), to_date("Last_Interaction_Date", "yyyy-MM-dd"))
)
display(silver_df.limit(10))
```

```python
silver_df.write.mode("overwrite").saveAsTable("patient_churn_prediction.patient_data.silver_patients")
```
