```python
import mlflow
import mlflow.spark
from pyspark.sql import functions as F

from pyspark.ml.functions import vector_to_array
```

```python
silver_patients_df = spark.read.table("patient_churn_prediction.patient_data.silver_patients")
display(silver_patients_df)
```

```python
# 1. Define your temporary path (Must be a Unity Catalog Volume)
# Ensure this Volume actually exists in Catalog Explorer
tmp_path = "/Volumes/patient_churn_prediction/log_model/log_model"

# 2. Load the model from the Registry
model_uri = "models:/patient_churn_prediction.default.patient_churn_logreg/4"

# We must specify dfs_tmpdir during the LOAD process as well on Serverless
loaded_model = mlflow.spark.load_model(model_uri, dfs_tmpdir=tmp_path)

# 3. Run Inference
predictions = loaded_model.transform(silver_patients_df)
```

```python
display(predictions.limit(10))
```

```python
# 1. Convert the probability vector into an array and extract the index [1]
predictions_with_prob = predictions.withColumn("prob_array", vector_to_array("probability")) \
                                   .withColumn("churn_probability", F.col("prob_array")[1])

# 2. Update your Gold Table selection
# If you want to keep Gender, enable schema overwrite
# Otherwise, remove Gender to match the table

gold_df = predictions_with_prob.select(
    "PatientID", 
    "Gender",
    "Age", 
    "State", 
    "Specialty", 
    "Overall_Satisfaction",
    F.round(F.col("churn_probability"), 4).alias("Churn_Score"), # The 0.8855 value
    F.col("prediction").alias("Churn_Predicted"),               # The 1 or 0 value
    F.when(F.col("churn_probability") > 0.90, "Very High Risk")
     .when((F.col("churn_probability") > 0.75) & (F.col("churn_probability") <= 0.90), "High Risk")
     .when((F.col("churn_probability") > 0.60) & (F.col("churn_probability") <= 0.75), "Moderate Risk")
     .when((F.col("churn_probability") > 0.55) & (F.col("churn_probability") <= 0.60), "Low Risk")
     .otherwise("Very Low Risk")
     .alias("Risk_Segment")
)

# 3. Save to Gold with schema overwrite
(
    gold_df.write
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("patient_churn_prediction.patient_data.gold_patients")
)

display(gold_df.limit(10))
```

```python
churn_score_desc_df = gold_df.select(
    F.min("Churn_Score").alias("Min_Churn_Score"),
    F.max("Churn_Score").alias("Max_Churn_Score"),
    F.avg("Churn_Score").alias("Avg_Churn_Score"),
    F.stddev("Churn_Score").alias("Stddev_Churn_Score"),
    F.expr("percentile(Churn_Score, array(0.25))")[0].alias("Q1_Churn_Score"),
    F.expr("percentile(Churn_Score, array(0.5))")[0].alias("Median_Churn_Score"),
    F.expr("percentile(Churn_Score, array(0.75))")[0].alias("Q3_Churn_Score")
)

display(churn_score_desc_df)
```

```python
churn_score_desc_df.write.mode("overwrite").saveAsTable("patient_churn_prediction.patient_data.churn_data")
```
